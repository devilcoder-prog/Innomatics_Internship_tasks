# Problem 1: Employee Performance Bonus Eligibility


```python

n = int(input("Enter number of employees: "))

d = {}

# input employee name and score
for i in range(n):
    name = input("Enter employee name: ")
    score = int(input("Enter performance score: "))
    d[name] = score

# find highest score
mx = max(d.values())

top = []

# store all employees with highest score
for name in d:
    if d[name] == mx:
        top.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top), f"(Score: {mx})")
```

# Problem 2: Search Query Keyword Analysis

```python
s = input("Enter search query: ").lower()

# remove punctuation
for ch in ".,!?;:":
    s = s.replace(ch, "")

w = s.split()
d = {}

# count frequency of each word
for x in w:
    if x in d:
        d[x] += 1
    else:
        d[x] = 1

res = {}

# store words searched more than once
for x in d:
    if d[x] > 1:
        res[x] = d[x]

print(res)
```

# Problem 3: Sensor Data Validation

```python

n = int(input("Enter number of readings: "))

a = []

# input sensor readings
for i in range(n):
    x = int(input("Enter reading: "))
    a.append(x)

v = []

# store even readings as (index, value)
for i in range(n):
    if a[i] % 2 == 0:
        v.append((i, a[i]))

print("Valid Readings (Hour, Value):")
print(v)
```

# Problem 4: Email Domain Usage Analysis

```python
n = int(input("Enter number of emails: "))

a = []

# input email ids
for i in range(n):
    e = input("Enter email: ")
    a.append(e)

d = {}

# count domain frequency
for e in a:
    dom = e.split("@")[1]
    if dom in d:
        d[dom] += 1
    else:
        d[dom] = 1

# print domain percentage
for dom in d:
    per = (d[dom] / n) * 100
    print(f"{dom}: {int(per)}%")
```

# Problem 5: Sales Spike Detection

```python
n = int(input("Enter number of days: "))

a = []

# input sales values
for i in range(n):
    x = int(input("Enter sales: "))
    a.append(x)

# calculate average
s = 0
for x in a:
    s += x

avg = s / n

# find spike days (more than 30% above average)
for i in range(n):
    if a[i] > avg * 1.3:
        print(f"Day {i+1}: {a[i]}")
```

# Problem 6: Duplicate User ID Detection


```python
n = int(input("Enter number of user ids: "))

a = []

# input user ids
for i in range(n):
    x = input("Enter user id: ")
    a.append(x)

d = {}

# count frequency
for x in a:
    if x in d:
        d[x] += 1
    else:
        d[x] = 1

# print only duplicates
for x in d:
    if d[x] > 1:
        print(f"{x} -> {d[x]} times")
```
